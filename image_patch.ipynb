```python
from pathlib import Path

import numpy as np
from patchify import patchify

from utility.plot import plot_image, plot_mask
from utility.preprocess import extract_patches, image_to_nparray
```

```python
# add the image and mask path
image_path = Path("./images/raster_1.tif")
mask_path = Path("./images/mask_1.tif")
# convert the image and mask into numpy array
images, masks = image_to_nparray(image_path, mask_path, 4)
# slice the image into patches by given resolution
images = patchify(images, (128, 128, 4), 128)
masks = patchify(masks, (128, 128, 1), 128)
# apply transformations and extract the images and masks
image, mask = extract_patches(images, masks, 10.0, 70.0)
# save the images and masks as numpy array
np.save('./image1.npy',image)
np.save('./label1.npy',mask)
```

```python
# randomly plot images and masks
i = np.random.randint(0, len(image))
plot_image(image,i)
plot_mask(mask,i)
```
